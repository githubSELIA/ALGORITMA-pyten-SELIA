```python
jumlah_barang = int(input("jumlah barang yang dibeli:"))
total_belanja = 0
for i in range(jumlah_barang):
  print(f"\nbarang ke-{i+1}:")
  harga = int(input("harga satuan barang: Rp"))
  jumlah = int(input("jumlah unit barang "))

  total_harga = harga * jumlah
  print(f" total harga barang ke-{i+1}: Rp{total_harga:.2f}")

  total_belanja += total_harga

  print(f"\n=============================================")
  print(f"total belanja keseluruhan: Rp{total_belanja:.2f}")
```
